```python
import time
import re
from datetime import datetime
```

```python
def scrape_bbc6_episode(url):
    # Set up Chrome options
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    
    # Initialize the driver
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(url)
    
    try:
        # Wait for the page to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "segments-list__item"))
        )
        
        # Click "Show more" button if it exists to reveal all tracks
        try:
            show_more = driver.find_element(By.CLASS_NAME, "ml__label--more")
            driver.execute_script("arguments[0].click();", show_more)
            time.sleep(2)  # Wait for animation to complete
        except:
            pass  # No "Show more" button or already expanded
        
        # Get DJ name
        dj_name = driver.find_element(By.CSS_SELECTOR, "a.context__item").text
        
        # # Get episode title
        # episode_title = driver.find_element(By.CSS_SELECTOR, "h1.no-margin").text
        
        # Extract episode date
        episode_date = None
        try:
            # Look for the span with title containing the date
            meta_span = driver.find_element(By.CSS_SELECTOR, "p.episode-panel__meta span[title]")
            date_str = meta_span.get_attribute("title")
            
            # Extract date using regex
            date_match = re.search(r'(\w{3} \d{1,2} \w+ \d{4})', date_str)
            if date_match:
                date_str = date_match.group(1)
                # Parse the date
                episode_date = datetime.strptime(date_str, "%a %d %B %Y").strftime("%Y-%m-%d")
            else:
                # Try alternative date format in the broadcast event
                broadcast_date = driver.find_element(By.CSS_SELECTOR, ".broadcast-event__time[title]")
                if broadcast_date:
                    date_str = broadcast_date.get_attribute("title")
                    episode_date = datetime.strptime(date_str, "%d %b %Y").strftime("%Y-%m-%d")
        except Exception as e:
            print(f"Error extracting date: {e}")
            # Fallback to URL to extract episode ID and date if possible
            try:
                episode_id = url.split('/')[-1]
                # If episode ID contains date format (common in BBC URLs)
                # You might need to adjust this regex based on actual URL patterns
                date_match = re.search(r'(\d{8})', episode_id)
                if date_match:
                    date_str = date_match.group(1)
                    episode_date = f"{date_str[:4]}-{date_str[4:6]}-{date_str[6:8]}"
            except:
                episode_date = "Unknown Date"
        
        # Extract all music segments
        music_segments = driver.find_elements(By.CSS_SELECTOR, "li.segments-list__item--music")
        
        tracks = []
        for index, segment in enumerate(music_segments, 1):
            try:
                # Extract artist
                artist_elem = segment.find_element(By.CLASS_NAME, "artist")
                artist = artist_elem.text if artist_elem else "Unknown Artist"
                
                # Extract track title
                title_container = segment.find_element(By.CSS_SELECTOR, "p.no-margin")
                title_span = title_container.find_element(By.TAG_NAME, "span")
                title = title_span.text if title_span else "Unknown Title"
                
                # Add to our list
                tracks.append({
                    'dj': dj_name,
                    'episode_date': episode_date,
                    'track_order': index,
                    'artist': artist,
                    'title': title,
                })
            except Exception as e:
                print(f"Error extracting track {index}: {e}")
                continue
        
        return pd.DataFrame(tracks)
        
    finally:
        driver.quit()
```
